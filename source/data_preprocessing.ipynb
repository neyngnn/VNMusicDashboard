# Tiền xử lý dữ liệu

```python
import pandas as pd

# Đọc dữ liệu từ file data.csv
df = pd.read_csv('Data.csv')

df.info()
```

```python
# Xóa các dòng trùng lặp
df_cleaned = df.drop_duplicates()


# Lưu dữ liệu đã làm sạch vào file mới data_spotify.csv
df_cleaned.to_csv('data_spotify.csv', index=False)
```

```python
df_cleaned
```

```python
# Lấy các dữ liệu về thông số âm nhạc của bài hát
parameter_song = ['track_name', 'danceability', 'energy', 'acousticness', 'liveness', 'valence']
df_selected = df[parameter_song ]
df_selected 
```

```python
# Chuyển đổi dữ liệu từ dạng chiều ngang sang dạng theo chiều dọc
df_melted = df_selected.melt(id_vars=['track_name'], var_name='index', value_name='value')
df_melted 
```

```python
# Lưu dữ liệu vào file mới trackname.csv
df_melted.to_csv('trackname.csv', index=False)
```
